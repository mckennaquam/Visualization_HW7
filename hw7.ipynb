```python
import pandas as pd
import altair as alt
```

```python
#amsterdam 
df_amsterdam_weekdays = pd.read_csv("amsterdam_weekdays.csv")
df_amsterdam_weekends = pd.read_csv("amsterdam_weekends.csv")

#athens
df_athens_weekdays = pd.read_csv("athens_weekdays.csv")
df_athens_weekends = pd.read_csv("athens_weekends.csv")

#barcelona
df_barcelona_weekdays = pd.read_csv("barcelona_weekdays.csv")
df_barcelona_weekends = pd.read_csv("barcelona_weekends.csv")

#london
df_london_weekdays = pd.read_csv("london_weekdays.csv")
df_london_weekends = pd.read_csv("london_weekends.csv")

#paris
df_paris_weekdays = pd.read_csv("paris_weekdays.csv")
df_paris_weekends = pd.read_csv("paris_weekends.csv")

#rome
df_rome_weekdays = pd.read_csv("rome_weekdays.csv")
df_rome_weekends = pd.read_csv("rome_weekends.csv")
```

data notes
amsterdam: biggest differce between weekday and weekend, cheaper weakday (need to confirm), amsterdam most bedrooms, most expeive
athens: Cleanliness, Satisfaction, capacity, cheapest 
barcelona: lowest capacity
london: lowest satisfaction
paris: lowest bedroom, lowest metro distance, cheaper weekend
rome: most superhost

```python
data = [["Amsterdam", "Weekdays", round(df_amsterdam_weekdays["realSum"].mean(), 2)],
       ["Amsterdam", "Weekends", round(df_amsterdam_weekends["realSum"].mean(), 2)],
       ["Athens", "Weekdays", round(df_athens_weekdays["realSum"].mean(), 2)],
       ["Athens", "Weekends", round(df_athens_weekends["realSum"].mean(), 2)],
       ["Barcelona", "Weekdays", round(df_barcelona_weekdays["realSum"].mean(), 2)],
       ["Barcelona", "Weekends", round(df_barcelona_weekends["realSum"].mean(), 2)],
       ["London", "Weekdays", round(df_london_weekdays["realSum"].mean(), 2)],
       ["London", "Weekends", round(df_london_weekends["realSum"].mean(), 2)],
       ["Paris", "Weekdays", round(df_paris_weekdays["realSum"].mean(), 2)],
       ["Paris", "Weekends", round(df_paris_weekends["realSum"].mean(), 2)],
       ["Rome", "Weekdays", round(df_rome_weekdays["realSum"].mean(), 2)],
       ["Rome", "Weekends", round(df_rome_weekends["realSum"].mean(), 2)]]

df_format = pd.DataFrame(data, columns=['City', 'Period', "Average_Price"])
df_format
```

```python
df_format = df_format.sort_values(by=["Average_Price"])

weekday =  alt.Chart(df_format).mark_().encode(
    x = "City",
    y = "Average_Price",
    color = alt.value("blue")
).transform_filter(
    alt.FieldEqualPredicate(field='Period', equal="Weekdays")
)

weekdend =  alt.Chart(df_format).mark_line().encode(
    x = "City",
    y = "Average_Price",
    color = alt.value("red")
).transform_filter(
    alt.FieldEqualPredicate(field='Period', equal="Weekends")
)
weekdend + weekday
```

```python
alt.Chart(df_price).mark_bar().encode(
    x = alt.X("City", sort="-y", axis=alt.Axis(labelAngle=-45)),
    y = alt.Y("price_avg", axis=alt.Axis(title="Average Price (USD$)")),
    color = alt.value("#ff625e")
).properties(
    width =250,
    height = 350
)
```

```python
# real sum is the price for 2 people for 2 nights so the price per person per night would be the real sum divied by 4
df_amsterdam_weekdays["price_per_person"] = df_amsterdam_weekdays["realSum"] / 4
df_athens_weekdays["price_per_person"] = df_athens_weekdays["realSum"] / 4
df_barcelona_weekdays["price_per_person"] = df_barcelona_weekdays["realSum"] / 4
df_london_weekdays["price_per_person"] = df_london_weekdays["realSum"] / 4
df_paris_weekdays["price_per_person"] = df_paris_weekdays["realSum"] / 4
df_rome_weekdays["price_per_person"] = df_rome_weekdays["realSum"] / 4

data_price_per_person = [["Amsterdam", df_amsterdam_weekdays['price_per_person'].mean()],
                  ["Athens", df_athens_weekdays['price_per_person'].mean()],
                  ["Barcelona", df_barcelona_weekdays['price_per_person'].mean()],
                  ["London", df_london_weekdays['price_per_person'].mean()],
                  ["Paris", df_paris_weekdays['price_per_person'].mean()],
                  ["Rome", df_rome_weekdays['price_per_person'].mean()]]

df_price_per_person = pd.DataFrame(data_price_per_person, columns=['City', "Price per person"])
df_price_per_person = df_price_per_person.sort_values(by=['Price per person'], ascending=False)
df_price_per_person
```

```python
alt.Chart(df_price_per_person).mark_bar().encode(
    x = alt.X("City", sort="-y", axis=alt.Axis(labelAngle=-45)),
    y = alt.Y("Price per person", axis=alt.Axis(title="Price per person (USD$)")),
    color = alt.value("#ff526e")
).properties(
    width =250,
    height = 350,
    title="Price per person for one night"
)
```
